# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#### Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in data analysis and scientific computing. It provides a wide range of functions and tools for creating various types of plots, graphs, and charts to visualize data in a clear and informative way. Matplotlib is highly customizable, making it suitable for various applications, from basic exploratory data analysis to publication-quality figures in research papers.

#### Matplotlib is widely used for the following reasons:
1. Data Visualization: Matplotlib allows users to create a wide variety of static and interactive visualizations, making it easier to understand complex datasets and patterns.
2. Ease of Use: The library is user-friendly and straightforward, making it accessible to beginners while offering advanced customization options for more experienced users.
3. Integration with NumPy: Matplotlib integrates well with NumPy, another popular Python library for numerical computing, which simplifies the handling and plotting of numerical data.
4. Compatibility: Matplotlib is compatible with various operating systems and can create output in different formats, such as images (PNG, JPEG), PDF, and vector graphics (SVG).
5. Publication-Quality Figures: Researchers and scientists often use Matplotlib to create high-quality, publication-ready figures for scientific publications and presentations.


#### Five plots that can be plotted using Pyplot module of Matplotlib:
1. Line Plot: A line plot is used to display data points connected by straight lines. It is commonly used to visualize the trend or pattern in a dataset over time or continuous intervals.
2. Scatter Plot: A scatter plot displays individual data points as dots on a 2D plane. It is useful for visualizing the relationship between two variables and identifying patterns or correlations.
3. Bar Plot: A bar plot presents categorical data using rectangular bars, where the length of each bar corresponds to the value of the category. It is ideal for comparing discrete categories or showing frequency distributions.
4. Histogram: A histogram represents the distribution of continuous numerical data by dividing the data into intervals (bins) and displaying the frequency or count of data points falling into each bin.
5. Pie Chart: A pie chart is a circular chart that shows the proportion or percentage of different categories relative to the whole. It is useful for illustrating parts of a whole and comparing relative sizes of different components.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#### import numpy as np
#### np.random.seed(3)
#### x = 3 + np.random.normal(0, 2, 50)
#### y = 3 + np.random.normal(0, 2, len(x))

#### Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#### Use the following data:
##### import numpy as np
##### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
##### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
##### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
##### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### The subplot() function in Matplotlib is used to create multiple plots (subplots) within the same figure. It allows you to divide the figure into a grid of rows and columns, and then specify which subplot you want to plot your data in. This is useful when we want to visualize multiple datasets or different aspects of the same dataset side by side for comparison or to maintain a clean layout in our visualizations.

#### The syntax of the subplot() function is plt.subplot(nrows, ncols, index), where nrows is the number of rows in the grid, ncols is the number of columns in the grid, and index specifies the position of the subplot in the grid, starting from 1 (left-to-right, top-to-bottom).

```python
import numpy as np
import matplotlib.pyplot as plt

#For Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

#for Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

#for Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

#for Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2,2,1)
plt.plot(x1,y1,label='Line 1')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot 1")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label="Line 2")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot 2")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label="Line 3")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot 3")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x4, y4, label="Line 4")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot 4")
plt.legend()

plt.tight_layout()

plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
##### import numpy as np
##### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
##### profit = np.array([3000, 8000, 1000, 10000])

#### A bar plot is a type of data visualization that represents categorical data using rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots are commonly used to compare the values of different categories or to show the frequency distribution of categorical data.

#### Bar plots are used for the following reasons:
1. Comparison of Categories: Bar plots are effective for comparing the values of different categories. They make it easy to see which category has the highest or lowest value.
2. Visualizing Frequency Distributions: In the case of categorical data, bar plots can display the frequency or count of each category, making it easy to identify the most common or least common categories.
3. Visualizing Trends over Time: Bar plots can be used to show how the values of categories change over time, especially when using stacked or grouped bar plots.
4. Presenting Categorical Data: Bar plots are a popular choice for presenting survey results, market share data, and other forms of categorical data in a visually appealing and straightforward manner.
5. Comparing Multiple Groups: Bar plots can be used to compare data across multiple groups simultaneously, such as comparing the performance of different companies, products, or teams.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit,label="bar plot")
plt.xlabel("company")
plt.ylabel("profit")
plt.title("bar plot")
plt.legend()
plt.show()
```

```python
plt.barh(company,profit,label="horizontal bar plot")
plt.xlabel("company")
plt.ylabel("profit")
plt.title("horizontal bar plot")
plt.legend()
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
##### box1 = np.random.normal(100, 10, 200)
##### box2 = np.random.normal(90, 20, 200)

#### A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution and summary statistics of a dataset. It provides a visual summary of the data's central tendency, dispersion, and potential outliers. The key components of a box plot are the "box," which represents the interquartile range (IQR) and median, and the "whiskers," which extend to the minimum and maximum values within a specified range.

#### How is a box plot constructed:
1. The box plot consists of a rectangular "box" that represents the IQR, which is the range between the first quartile (Q1) and the third quartile (Q3). The median (Q2) is indicated by a line inside the box.
2. The "whiskers" extend from the edges of the box to the minimum and maximum values within a specified range, often determined by a set percentage of the IQR. Points beyond the whiskers are considered potential outliers.
3. Outliers, if present, are plotted as individual data points beyond the whiskers.

#### Box plots are used for the following reasons:
1. Visualizing Data Distribution: Box plots provide a concise and informative visualization of the distribution of data, allowing for easy comparison between different groups or datasets.
2. Identifying Central Tendency: The box represents the IQR and median, giving insights into the central tendency of the data. The median provides a measure of the data's central location, and the IQR shows the spread of the middle 50% of the data.
3. Detecting Outliers: Box plots help in identifying potential outliers in the dataset, which are data points that fall significantly above or below the majority of the data.
4. Comparing Distributions: Box plots are valuable for comparing the distributions of different datasets, especially when multiple box plots are placed side by side.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data= [box1,box2]
labels =['Box 1','Box 2']

plt.figure(figsize=(8,6))
plt.boxplot(data, labels=labels,vert=True, patch_artist=True)


plt.xlabel("Box")
plt.ylabel("Value")

plt.title("Box Plot of Box 1 and Box 2")
plt.show()
```
