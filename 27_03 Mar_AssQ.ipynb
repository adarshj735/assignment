# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Seaborn offers a variety of specialized plots that are particularly useful for visualizing different aspects of data distribution and relationships. Here are five common plots you can create using Seaborn, along with explanations of their uses:

#### Histogram:

#### Description: A histogram is a graphical representation of the distribution of a dataset. It divides the data into "bins" and displays the frequency or count of data points within each bin.
#### Use: Histograms help us understand the distribution of a continuous variable. They are useful for identifying patterns such as skewness, central tendency, and outliers. Histograms are commonly used in exploratory data analysis to gain insights into the data's characteristics.

#### Scatter Plot:
#### Description: A scatter plot displays individual data points as dots on a two-dimensional plane. Each point's position is determined by the values of two variables.
#### Use: Scatter plots are used to visualize the relationship between two continuous variables. They help identify trends, correlations, clusters, and outliers. Scatter plots are essential for understanding the degree and nature of the relationship between variables.

#### Pair Plot:
#### Description: A pair plot, also known as a scatterplot matrix, shows scatter plots of all pairs of variables in a dataset. The diagonal typically displays histograms or kernel density estimates.
#### Use: Pair plots are excellent for quickly exploring pairwise relationships between multiple variables. They help us identify potential correlations and patterns across the entire dataset, making them valuable in initial data exploration.

#### Box Plot (Box-and-Whisker Plot):
#### Description: A box plot summarizes the distribution of a continuous variable by showing the median, quartiles, and potential outliers. The "box" represents the interquartile range (IQR), and the "whiskers" extend to show the spread of the data.
#### Use: Box plots are used to visualize the spread, skewness, and potential outliers of a variable. They are especially useful for comparing the distributions of different groups or categories in a dataset, helping us understand the variability and central tendency of the data.

#### Violin Plot:
#### Description: A violin plot combines elements of a box plot and a kernel density plot. It displays the distribution of data across different levels or categories as a symmetrical "violin" shape.
#### Use: Violin plots are effective for comparing the distribution of data across different groups or categories. They provide information about the median, quartiles, and the underlying data density. Violin plots are particularly useful when we want to show the shape of the distribution along with summary statistics.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
sns.lineplot(x=fmri.timepoint,y=fmri.signal, hue=fmri.event,style=fmri.region)
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal by Timepoint for Different Events and Regions")
plt.legend()
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'

```python
titanic=sns.load_dataset('titanic')

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(x=titanic.pclass,y=titanic.age)
plt.xlabel("Pclass")
plt.ylabel("Age")
plt.title("Box plot of Age vs Pclass")

plt.subplot(1,2,2)
sns.boxplot(x=titanic.pclass,y=titanic.fare)
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.title("Box plot of Titanic Fare vs Pclass")

plt.tight_layout()

plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
diamond=sns.load_dataset('diamonds')

plt.figure(figsize=(12,6))
sns.histplot(x=diamond.price, hue= diamond.cut, multiple="stack")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram of Diamond Prices by Cut")


plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
iris= sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flight=sns.load_dataset("flights")
flight_piv=flight.pivot_table(index='month',columns='year',values='passengers')
plt.figure(figsize=(10,6))
sns.heatmap(flight_piv,annot=True,fmt='d',cmap='RdPu')
plt.xlabel('Year')
plt.ylabel('Month')
plt.title("Flights Dataset Heatmap")
plt.show()
```
